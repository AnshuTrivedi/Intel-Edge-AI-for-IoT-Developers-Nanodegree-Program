We've seen now that a key characteristic of FPGAs is that they are field-programmable.
But how do we actually program the FPGA's circuitry in practice? It turns out that there are a number of different ways we can approach this, at different levels of abstraction. Let's have a look.

## Developing program at different levels of abstraction

* Different levels of abstraction
* Register Transfer Level(RTL)

We can do programming at different level of abstraction,lowest level of abstraction is `Register Transfer Level(RTL)`.It invovles specifying configuration of actual hardware registers. This can be done most directly using `Register Transfer Language.`
* Register Transfer Language(RTL)
Once the circuits have been designed at Register Transfer Level,we use this design to run through routning and placement tool and affect the hardware compiler to generate the bitstream.Feed binary sequence into FPGA to do actual configurations of logic blocks and interconnects

```python
from IPython.display import Image
Image(filename='diff_prog_fpga.png')
```

Again, notice that the three lowest levels of abstraction all ultimately generate **`Register Transfer Level (RTL) output`**, which is then used to create a **`bitstream`** (and it is this bitstream that is fed into the FPGA to do the actual configuration of the logic blocks and interconnects). In contrast, OpenVINO leverages bitstreams that are pre-compiled and ready to be loaded at runtime.

## Converting an FPGA into an AI Accelerator
One of the advantages of an ASIC is that it is designed to carry out a particular type of task with optimal performance. The downside is that the ASIC is then optimized for that particular type of task, but does not perform as well (if at all) on other tasks.

In contrast, `FPGAs can be reprogrammed to optimize their performance for different functions as needed`. Thus, we can program an FPGA to act as an AI accelerator so that it performs well when running inference. However, we don't want to program it as a general AI accelerator—rather, we want to program it as an accelerator for the particular model we are going to run.
